```python
dummy_train_df = all_dummy_df.loc[:1460]
dummy_test_df = all_dummy_df.loc[1461:]
```

```python
dummy_train_df=pd.concat((dummy_train_df,y_train),axis=1)
dummy_train_df
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X=dummy_train_df.drop('SalePrice',axis=1)
y=dummy_train_df.SalePrice
```

```python
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.3,
                max_depth = 5, alpha = 10, n_estimators = 100)
xg_reg.fit(X_train,y_train)
y_preds = xg_reg.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_preds))
print("RMSE: %f" % (rmse))
print('R2 score: {}'.format(r2_score(y_test, y_preds)))
```

```python
print(y_preds.shape)
print(y_test.shape)
```

```python
xg_reg.fit(X_train, y_train, eval_metric='rmse', verbose = True, eval_set = [(X_test, y_test)],early_stopping_rounds=50)
preds = xg_reg.predict(X_test)
```
